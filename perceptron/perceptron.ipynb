```python
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import numpy as np
import time
from IPython import display
pylab.rcParams['figure.figsize'] = 10,10
```

```python
class gaussian_data:
    def __init__(self, dimensions=1, center=[0], covariance=[1]):
        self.dimensions = dimensions
        if np.shape(covariance)[0] != dimensions:
            covariance = np.identity(dimensions)
        if np.shape(center)[0] != dimensions:
            center = np.zeros(dimensions)
        self.center = center
        self.covariance = covariance
    
    def generate(self, N=1):
        points = np.random.normal(loc=self.center, size=(N, self.dimensions))
        return np.dot(points,self.covariance)
```

```python
def SSE(Y, W, X):
    """
    Sum of Squared Error:
        Y = column vector target
        W = column vector weights
        X = matrix inputs
    Returns:
            sse = 0
            for i, y in enumerate(Y):
                sse += (y - np.dot(W, np.append(X[i,:], 1)))**2
            return sse
    """
    return np.sum((Y.T - np.dot(W.T, np.c_[X, np.ones(X.shape[0]).T].T))**2)
```

```python
def MSE(Y, W, X):
    """
    Mean Squared Error:
        Y = column vector target
        W = column vector weights
        X = matrix inputs
    Returns:
            sse = 0
            for i, y in enumerate(Y):
                sse += (y - np.dot(W, np.append(X[i,:], 1)))**2
            return sse/X.shape[0]
    """
    return SSE(Y, W, X)/X.shape[0]
```

```python
def accuracy(T, W, X):
    y = np.dot(W.T, np.c_[X, np.ones(X.shape[0]).T].T)
    y = (y > 0)
    t = (T > 0)
    return np.sum(y == t)/np.float(T.shape[0])
    
```

```python
def plot_points_and_lines(X,T,W=None, highlight=None):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.scatter(X[:,0], X[:,1], c=T, s=60, cmap=plt.cm.hot)
    ax.plot([-100,100], [0,0], c='red')
    ax.plot([0,0], [-100,100], c='red')
    
    colors = np.copy(T)
    if highlight != None:
        colors[highlight] = 0
        
    # Origins
    Xo = Yo = np.zeros(np.shape(X)[0])

    ax.quiver(Xo,Yo,X[:,0],X[:,1],colors, angles='xy',scale_units='xy',scale=1, 
              linewidths=(1,), edgecolors=('k'), headaxislength=5, cmap=plt.cm.hot)


    # Draw hyperplanes of class 0
    class_0 = (T == -1)
    hyperplanes_o = np.c_[-X[class_0,1], X[class_0,0]]*100
    hyperplanes_e = -hyperplanes_o
    ax.plot([hyperplanes_o[:,0], hyperplanes_e[:,0]], 
            [hyperplanes_o[:,1], hyperplanes_e[:,1]], '-', c='gray')

    # Draw hyperplanes of class 1
    class_1 = (T == 1)
    hyperplanes_o = np.c_[-X[class_1,1], X[class_1,0]]*100
    hyperplanes_e = -hyperplanes_o
    ax.plot([hyperplanes_o[:,0], hyperplanes_e[:,0]], 
            [hyperplanes_o[:,1], hyperplanes_e[:,1]], '-.', c='black')
    
        
    if W != None:
        ax.scatter(W[0], W[1], s=60, c='green')
        ax.quiver(0,0,W[0],W[1], angles='xy',scale_units='xy',scale=1, 
              linewidths=(2,), edgecolors=('green'), color='green', headaxislength=5)
        ax.plot([-100*W[1], 100*W[1]], [100*W[0], -100*W[0]], '-', c='green', linewidth=2)

    x_max = y_max = np.max(np.abs(X))   
    x_min = y_min = -x_max
    ax.set_xlim([x_min, x_max])
    ax.set_ylim([y_min, y_max])
    return fig, ax
```

```python
def plot_points(X,T,W=None, highlight=None):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    
    colors = np.copy(T)
    if highlight != None:
        colors[highlight] = 0
        
    ax.scatter(X[:,0], X[:,1], c=colors, s=60, cmap=plt.cm.hot)
    ax.plot([-100,100], [0,0], c='red')
    ax.plot([0,0], [-100,100], c='red')

    if W != None:
        ax.scatter(W[0], W[1], s=60, c='green')
        ax.quiver(0,0,W[0],W[1], angles='xy',scale_units='xy',scale=1, 
              linewidths=(2,), edgecolors=('green'), color='green', headaxislength=5)
        ax.plot([-100*W[1], 100*W[1]], [100*W[0], -100*W[0]], '-', c='green', linewidth=2)

    x_max = y_max = np.max(np.abs(X))   
    x_min = y_min = -x_max
    ax.set_xlim([x_min, x_max])
    ax.set_ylim([y_min, y_max])
    return fig, ax
```

```python
N1 = 10
DIM = 1
C1 = gaussian_data(dimensions=DIM, center=[0], covariance=[1])
X1 = C1.generate(N1)

N2 = 50
C2 = gaussian_data(dimensions=DIM, center=[5], covariance=[1])
X2 = C2.generate(N2)

# Input
X = np.concatenate((X1, X2))
# Target
T = np.concatenate((np.ones(N1)*-1, np.ones(N2)))

index = np.random.permutation(T.shape[0])
X = X[index]
T = T[index]

w = np.random.normal(loc=0, scale=1, size=(DIM+1))
print w
```

```python
plt.scatter(X, T, c=T, cmap=plt.cm.Set3)
xx = np.linspace(np.min(X), np.max(X), 2)
z = (-w[0]*xx - w[1])
    
plt.plot(xx,z)
```

```python
alpha = 0.2
acc = np.ones(X.shape[0])
for i, sample in enumerate(X):
    sample = np.append(sample,1).T
    y = np.dot(w.T, sample)
    if y < 0:
        y = -1.
    else:
        y = 1.
    w = w + alpha*(T[i] - y)*sample
    acc[i] = accuracy(T,w,X)
    plt.clf()
    plt.scatter(X, T, c=T, cmap=plt.cm.Set3)
    
    xx = np.linspace(np.min(X), np.max(X), 2)
    z = (-w[0]*xx - w[1])
    
    plt.plot(xx,z)
    
    display.clear_output(wait=True)
    display.display(plt.gcf())
    time.sleep(.2)
```
